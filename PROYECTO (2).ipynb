```python
def estatus_cred(row):
    if row["loan_status"] in ["Charged Off", "Default"]:
        return "BROKEN"
    elif row["loan_status"] == "Fully Paid":
        return "COMPLETE"
    else:
        return "ACTIVE"
```

```python
dataset["settlement_status"] = dataset.apply(estatus_cred, axis=1)
```

### 1. Definicion de mal pagador 

###### -Eliminar los nulos de last_pyment_d y next_pymnt_d
###### - Transformar la columna last_pymnt_d  y next_pymnt_d en formato fecha 
###### - crear la definicion donde la diferencia de dias entre el ultimo mes de pago y el proximo mes de pago no excedan los 45 días, caso contrario sería un mal pagador. 

```python
ult_pago= dataset.dropna(subset=['last_pymnt_d'])
```

```python
prox_pago= dataset.dropna(subset=['next_pymnt_d'])
```

```python
dataset.last_pymnt_d.unique()
```

```python
ult_pago['last_pymnt_d'] = pd.to_datetime(ult_pago['last_pymnt_d'])
```

```python
prox_pago['next_pymnt_d'] = pd.to_datetime(prox_pago['next_pymnt_d'], format="%b-%y")
```

```python
dataset['late_pymnt'] = (prox_pago['next_pymnt_d']- ult_pago['last_pymnt_d']).dt.days
```

```python
##Funcion que define el buen pagador y mal pagador 
```

```python
def late_pymnt_d (row):
    if row["late_pymnt"] > 45:
        return "Mal pagador"
    else:
        return "Buen pagador"
```

```python
dataset["costumer_status"] =dataset.apply(late_pymnt_d, axis=1)
```

### Feature transformation


##### 1. En este apartado usted deberá crear nuevas columnas apartir de las que tiene. Pueden ser:

```python
## Nueva columna de la classe del cliente doonde se define una funcion con parametros de la categoria crediticia del cliente
def customer_class(row):
    if row['late_pymnt'] <= 45:
        return "A"
    elif row['late_pymnt'] > 45 and row['late_pymnt'] <= 60:
        return "B"
    else:
        return "C"
    
```

```python
dataset["custmr_class"]=dataset.apply( customer_class, axis=1)
```

```python
## Calcular el ratio de clientes por estado crediticio
clientes_C = dataset["custmr_class"].value_counts()[0]
clientes_A = dataset["custmr_class"].value_counts()[1]
clientes_B = dataset["custmr_class"].value_counts()[2]
```

```python
total_clientes = len(dataset)

ratio_A = clientes_A / total_clientes
ratio_B = clientes_B / total_clientes
ratio_C = clientes_C / total_clientes
```

```python
def rat_cost_class (row):
    if row["custmr_class"]== "A":
        return ratio_A
    elif row["custmr_class"]== "B":
        return ratio_B
    else:
        return ratio_C
        
```

```python
dataset["ratio_cred"]=dataset.apply(rat_cost_class, axis=1)
```
